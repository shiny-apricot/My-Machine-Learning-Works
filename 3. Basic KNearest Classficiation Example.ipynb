```python
import numpy as np
import pandas as pd

data = pd.read_csv("car.data")

data.head()
```

```python
len(data)
```

```python
from sklearn import preprocessing as prep

encoder = prep.LabelEncoder()

prep_data = data.apply(encoder.fit_transform)

y_data = prep_data["class"]
x_data = prep_data.drop(["class"], axis=1)
```

```python
prep_data
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=44)
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn_model = KNeighborsClassifier(n_neighbors=7)
kn_model.fit(x_train, y_train)

score = kn_model.score(x_test,y_test)
score
```

```python
predicted = kn_model.predict(x_test)
```

```python
names = ["unacc","acc", "good", "vgood"]

for x in range(len(predicted)):
    prd_x = names[predicted[x]]
    prd_y = names[y_test.iloc[x]]
    
    accuracy_text = ''
    if prd_x == prd_y:
        accuracy_text = 'TRUE'
    else:
        accuracy_text = 'FALSE =============='
    print("Predicted= ", prd_x, "Actual= ",prd_y, "      ", accuracy_text  )
```
